```python
import pandas as pd
import numpy as np
import os
import random
import time
import json
from pprint import pprint
```

```python
def get_tabular_results(input_dir, req_type, pearson_or_spearmann):
    if pearson_or_spearmann == "pearson":
        check = "Pearson Correlation = \n"
    else:
        check = "Spearmann Correlation = \n"
    data = {"Question": ["", "", "", "Q1", "Q2", "Q3", "Q4", "Q5", "Q6"]}
    
    for dirpath, dirs, files in os.walk(input_dir):
        dirname = dirpath.split("/")[-1]
        for filename in files:
#             print(filename)
            if filename.split(".")[0] == req_type:
                data[dirname + " Rho"] = ["_".join(req_type.split("_")[1:]), pearson_or_spearmann, "Rho"]
                data[dirname + " p"] = ["_".join(req_type.split("_")[1:]), pearson_or_spearmann, "p"]
                with open(os.path.join(dirpath, filename), 'r') as f1:
                    lines_here = f1.readlines()
#                     print(lines_here)
                    for i, line in enumerate(lines_here):
                        if line == check:
#                             print('ea')
                            for j in range(1, 7):
                                qx_res = lines_here[i + j]
                                qx_rho = qx_res.split(',')[0].split('= ')[1]
#                                 print(qx_rho)
                                qx_p = qx_res.split(',')[1].split('= ')[1]
#                                 print(dirname + " Rho")
                                data[dirname + " Rho"].append(qx_rho.strip())
#                                 print(data[dirname + " Rho"])
                                data[dirname + " p"].append(qx_p.strip())
#     print(data)
    df = pd.DataFrame(data)
    df = df.reindex(sorted(df.columns), axis=1)
    req_type_filename = "_".join(req_type.split("_")[1:])
    df.to_csv(os.path.join(input_dir, f"compiled_results_{req_type_filename}_{pearson_or_spearmann}.csv"), index=False)
    return df
```

```python
input_dir = "gpt2-large-finetuned-10-epochs-2e-05-lr-no-weight-decay-scheduler/"
for req_type in ["RESULTS_tc_test", "RESULTS_pc_test", "RESULTS_tc_train", "RESULTS_pc_train"]:
    for pearson_or_spearmann in ["pearson", "spearmann"]:
        df = get_tabular_results(input_dir, req_type, pearson_or_spearmann)
        pprint(df)
```
